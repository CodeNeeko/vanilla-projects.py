```python
import numpy as np
import matplotlib.pyplot as plt

def foo(X):
    return 4*X + 5

X = np.arange(1,11,1)
Y = foo(X)
plt.scatter(X,Y)

num = np.dot((X - np.mean(X)),(Y))

den = np.dot((X - np.mean(X)),(X - np.mean(X)))

b1 =num / den

b0 = np.mean(Y) - b1 * np.mean(X)
b1, b0
```
